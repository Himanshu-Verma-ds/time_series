```python
path = r'E:\Toolkit\vs_code_files\DATASETS\techniques_test_df.csv'

df = pd.read_csv(path)
df
```

```python
df.drop(columns= ['Unnamed: 0', 'Sales of Accessories - Income_y'], inplace= True)
df['period'] = pd.to_datetime(df['period'])

df.set_index(['period'], inplace= True)
df.head()
```

```python
len(df.columns)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
df_scaled = sc.fit_transform(df)
```

```python
df_scaled
```

```python
df_scaled = pd.DataFrame(df_scaled, columns= df.columns)
df_scaled.head()
```

```python
from sklearn.ensemble import RandomForestRegressor

X = df_scaled.drop(columns= ['Real GDP at Market Prices'])
y = df_scaled['Real GDP at Market Prices']

rf = RandomForestRegressor(n_estimators= 100, random_state= 1)
rf.fit(X,y)

importances = rf.feature_importances_

# Plot importances
import matplotlib.pyplot as plt

plt.bar(range(X.shape[1]), importances)
plt.xlabel('Feature Index') 
plt.ylabel('Feature Importance')
plt.show()
```
